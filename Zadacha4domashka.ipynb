```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://gbcdn.mrgcdn.ru/uploads/asset/5087678/attachment/a3a1af2058ad08de6ab63618c1dfeb9f.csv', sep=',')
df.head()
```

1.2 Изучите стоимости недвижимости

```python
plt.figure(figsize=(8, 6))
plt.hist(df['price'])
plt.xlabel('price')
plt.ylabel('Count')
plt.title('стоимости недвижимости');
```

1.3 Изучите распределение квадратуры жилой

```python
plt.hist(df['sqft_living'])
plt.xlabel('sqft_living')
plt.ylabel('Count')
plt.title('распределение квадратуры жилой');
```

1.4 Изучите распределение года постройки

```python
plt.hist(df['yr_built'])
plt.xlabel('yr_built')
plt.ylabel('Count')
plt.title('распределение года постройки');
```

2.1 Изучите распределение домов от наличия вида на набережную
Постройте график
Сделайте выводы

```python
x = df['waterfront'].value_counts()
x
plt.figure(figsize=(8, 5))
plt.pie(x.values, labels = x.index, autopct='%1.1f%%')
plt.legend(['no', 'yes'])
print(x);
```

Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix [np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
plt.figure(figsize=(10,8))
sns.set(font_scale=1.4)
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm')
plt.title('Correlation matrix');
```
